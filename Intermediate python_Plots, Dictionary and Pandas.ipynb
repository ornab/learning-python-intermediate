`````python
import numpy as np
import matplotlib.pyplot as plt

# Example data for FPR and TPR
fpr = [0.2, 0.2, 0.2, 0.2, 0.4, 0.6, 0.8, 1]
tpr = [0, 0.33, 0.67, 1, 1, 1, 1, 1]

# Calculate AUC by counting boxes
total_boxes = len(fpr) - 1  # Total number of boxes
boxes_under_curve = 0
for i in range(total_boxes):
    if tpr[i] != tpr[i+1]:
        boxes_under_curve += 1

# Plot the ROC curve with little boxes
plt.figure()
plt.plot(fpr, tpr, color='blue')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')

# Display little boxes
for i in range(total_boxes):
    if tpr[i] != tpr[i+1]:
        plt.fill_between([fpr[i], fpr[i+1]], [tpr[i], tpr[i]], alpha=0.3, color='green')
    else:
        plt.fill_between([fpr[i], fpr[i+1]], [tpr[i], tpr[i+1]], alpha=0.3, color='green')

# Calculate AUC
auc = boxes_under_curve / total_boxes
print("AUC: ", auc)

plt.show()
`````

`````python
import numpy as np
import matplotlib.pyplot as plt

# Example data for FPR and TPR
fpr = [0.2, 0.2, 0.2, 0.2, 0.4, 0.6, 0.8, 1]
tpr = [0, 0.33, 0.67, 1, 1, 1, 1, 1]

# Calculate partial AUC with specificity >= 0.7
specificity_threshold = 0.7
partial_auc = 0
for i in range(len(fpr) - 1):
    if 1 - fpr[i] >= specificity_threshold and 1 - fpr[i + 1] >= specificity_threshold:
        partial_auc += (tpr[i] + tpr[i + 1]) * (fpr[i + 1] - fpr[i]) / 2
    elif 1 - fpr[i] >= specificity_threshold and 1 - fpr[i + 1] < specificity_threshold:
        tpr_partial = tpr[i] + (specificity_threshold - (1 - fpr[i])) * (tpr[i + 1] - tpr[i]) / (fpr[i + 1] - fpr[i])
        partial_auc += (tpr[i] + tpr_partial) * (specificity_threshold - (1 - fpr[i])) / 2

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')

# Display partial AUC value
plt.text(0.5, 0.2, 'Partial AUC (Specificity >= 0.7) = %0.2f' % partial_auc,
         bbox=dict(facecolor='white', edgecolor='black'))

plt.show()
`````
